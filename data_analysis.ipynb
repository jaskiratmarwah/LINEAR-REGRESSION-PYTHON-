```
data=pd.read_csv("https://trello-attachments.s3.amazonaws.com/5d6a0bf6791d1e1097fdbd8d/5d6a0c1060068284fbaf3cfd/b800b0953983bcc0f24becac2e0e02fe/Housing_Modified_prepared.csv")
data.head(5)
```

```
data.corr()
```

```
import statsmodels.api as sm
sm.graphics.plot_corr(data.corr(), xnames=data.columns)
```

```
#create a model
import sklearn.linear_model as lm
lr=lm.LinearRegression()

#select the variables
Y=data["price"]  #dependent
ind_var=data.columns
X=ind_var.delete(0) #removes price
```

```
X=data[X]
X
```

```
lr.fit(X,Y)  #TRAIN THE LINEAR REGRESSION
print("coefficient", lr.coef_)
print("intercept", lr.intercept_)

#Y=m1X1+ m2X2 .... +mnXn + c
```

```

y_pred=lr.predict(X)
y_pred
data["pred_price"]=y_pred
data[["price","lotsize","pred_price"]]
```

```
dict1={}
for name in X.columns:
  var=int(input("ENTER"+name+":"))
  dict1[name]=var
userdata=pd.DataFrame(data=dict1,index=[0], columns=X.columns)
user_price= lr.predict(userdata)
print("predicted price is USD",user_price[0])  
  
```
